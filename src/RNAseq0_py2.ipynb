```python
"""
Avi Shah, 11/16/2022. Yang Lab NJMS

This script will update names for Mycobacterium
tuberculosis H37Rv samples.

"""

import os
import pandas as pd
import numpy as np

save_dir = os.path.join("/scratch", "as2654", "RNAseq0")
ipynb_dir = os.path.join("/home", "as2654", "Yang", "RNAseq0")
os.chdir(ipynb_dir)
print(os.listdir("."))
mtb_df_fn = "M.tubercolosis_counts.csv"
```

```python
# see what the index looks like
df = pd.read_csv(os.path.join(ipynb_dir, mtb_df_fn), sep=",", header=0, index_col=0)

rename0_map = {}
print(df.index.names)
print(len(df.index))
for gene_name_idx in range(len(df.index)):
    gene_name_orig = df.index[gene_name_idx]
    # keep unique tag last six digits
    gene_name_new = gene_name_orig[-6:]
    #print(gene_name_orig, gene_name_new)
    rename0_map[gene_name_orig] = gene_name_new

print(df.rename(index=rename0_map))
```

```python
# replace_out.txt was used to repair the 
# gbk submission in Chitale et al. 2022

# we will use this to get gene names and 
# legacy Rv#s

nmap = {}
nmap_fn = "replace_out.txt"
with open(nmap_fn, "r") as nmapf:
    for line in nmapf:
        elements = [k for k in line.split(" ") if k.strip()]
        if len(elements) > 1:
            nmap[elements[0][-6:]] = elements[1]
        elif elements:
            nmap[elements[0][-6:]] = elements[0]
        else:
            pass

for k, v in nmap.items():
    print(k, v)
```

```python
# perform renaming operations and save file
df_new = df.rename(index=rename0_map).rename(index=nmap)

sf = os.path.join(ipynb_dir, "M.tuberculosis_counts_mixedgenenames.csv")
df_new.to_csv(sf, index=True, header=True, sep=",")
```
